```python
import pandas as pd
import pathlib
```

```python
# current notebook directory
data_path = pathlib.Path().resolve().parent / "data"

df = pd.read_parquet(data_path / "valenbici_2025.parquet")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Year"] = df["fecha_actualizacion"].dt.year
df["Month"] = df["fecha_actualizacion"].dt.month
df["Day"] = df["fecha_actualizacion"].dt.day
df["Hour"] = df["fecha_actualizacion"].dt.hour
df["Minute"] = df["fecha_actualizacion"].dt.minute
```

```python
fechas = df["fecha_actualizacion"].dt.strftime("%Y-%m-%d--%H:%M")

df["unique_id"] = df["Direccion"].str.cat(fechas, sep="-")
```

```python
df.head()
```

```python
df["bikes_percent"] = df["Bicis_disponibles"] / df["Espacios_totales"] * 100
```

## Feature Engineering

```python
num_estaciones = df["Numero"].nunique()

df["bikes_percent_ma_4"] = df.groupby("Numero")["bikes_percent"].transform(
    lambda x: x.shift(1).rolling(window=4).mean()
)
df["bikes_percent_ma_12"] = df.groupby("Numero")["bikes_percent"].transform(
    lambda x: x.shift(1).rolling(window=12).mean()
)
df["bikes_percent_ma_96"] = df.groupby("Numero")["bikes_percent"].transform(
    lambda x: x.shift(1).rolling(window=96).mean()
)
```

## Training

```python
def split_train_test(df, test_days=7, train_days=30, date_col="fecha_actualizacion"):
    now = df[date_col].max()
    test_start_date = now - pd.Timedelta(days=test_days)
    test = df[df[date_col] >= test_start_date]
    train_start_date = test_start_date - pd.DateOffset(days=train_days)
    train = df[(df[date_col] >= train_start_date) & (df[date_col] < test_start_date)]
    return train, test


train, test = split_train_test(df)
```

```python
test
```

```python
from lightgbm import LGBMRegressor
```

```python
# Definir los parámetros para regresión cuantílica
params = {
    "objective": "quantile",
    "alpha": 0.05,  # límite inferior
    "verbosity": -1,
    "n_estimators": 200,
    "learning_rate": 0.01,
    "num_leaves": 31,
    "max_depth": -1,
    "random_state": 42,
}

features = [
    "bikes_percent_ma_4",
    "bikes_percent_ma_12",
    "bikes_percent_ma_96",
]

model_low = LGBMRegressor(**params)
params["alpha"] = 0.5  # mediana
model_med = LGBMRegressor(**params)
params["alpha"] = 0.95  # límite superior
model_up = LGBMRegressor(**params)

model_low.fit(train[features], train["bikes_percent"])
model_med.fit(train[features], train["bikes_percent"])
model_up.fit(train[features], train["bikes_percent"])
```

```python
# X_test = test[["Year", "Month", "Day", "Hour", "Minute"]]
X_test = test[features]
y_test = test["bikes_percent"]

# Predicciones
y_pred_low = model_low.predict(X_test)
y_pred_med = model_med.predict(X_test)
y_pred_up = model_up.predict(X_test)

# El intervalo de predicción al 95% sería:
# [y_pred_low, y_pred_up]
```

```python
y_pred_low[:5], y_pred_med[:5], y_pred_up[:5]
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
)
import numpy as np

print("Predicciones de regresión cuantílica:")
print("Mediana (50%):")
mae = mean_absolute_error(y_test, y_pred_med)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_med))

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# add predictions to test set
test["bikes_percent_pred"] = y_pred_med
test["bikes_percent_pred_low"] = y_pred_low
test["bikes_percent_pred_up"] = y_pred_up
```
